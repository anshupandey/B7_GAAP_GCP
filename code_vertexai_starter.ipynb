## Getting started with Vertex AI

```python
from google import genai
import os
client = genai.Client()
```

```python
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents="Tell me about key allys of Russia. Answer in 2 lines"
)
print(response.text)
```

```python
from google.genai import types
contents = [types.Content(role='user',parts=[types.Part(text="Explain quantum computing in 1 line")])]
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=contents
)
print(response.text)
```
